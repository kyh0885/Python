```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-4,4, 100)
y1 = 2**x
y2 = 3**x
y3 = 0.5**x

print('x = {}'.format(x))
print('y1 = {}'.format(y1))
print('y2 = {}'.format(y2))
print('y3 = {}'.format(y3))


plt.figure(figsize=(5,5))
plt.plot(x,y1, 'black', linewidth = 3, label = '$y=2^x$')
plt.plot(x,y2, 'cornflowerblue', linewidth = 3, label = '$y=3^x$')
plt.plot(x,y3, 'gray', linewidth = 3, label = '$y=0.5^x$')
plt.ylim(-2,6)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='lower right')
#plt.show()
plt.savefig('graph.png')
```
