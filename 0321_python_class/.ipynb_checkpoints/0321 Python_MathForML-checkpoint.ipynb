```python
import numpy as np
import matplotlib.pyplot as plt
#matplot inline

def f(w0,w1):
    return w0**2 + 2*w0*w1 + 3
def df_dw0(w0,w1):
    return 2*w0 + 2*w1
def df_dw1(w0,w1):
    return 2*w0 + 0*w1

w_range = 2
dw =0.25

w0 = np.arange(-w_range, w_range + dw, dw)
w1 = np.arange(-w_range, w_range + dw, dw)
wn = w0.shape[0]
ww0,ww1 = np.meshgrid(w0,w1) #(D)
ff=np.zeros((len(w0), len(w1)))
dff_dw0=np.zeros((len(w0), len(w1)))
dff_dw1=np.zeros((len(w0), len(w1)))
for i0 in range(wn):
    for i1 in range(wn):
        ff[i1, i0]= f(w0[i0],w1[i1])
        dff_dw0[i1, i0]= df_dw0(w0[i0],w1[i1])
        dff_dw1[i1, i0]= df_dw1(w0[i0],w1[i1])
    
plt.figure(figsize=(9,4))
plt.subplots_adjust(wspace = 0.3)
plt.subplot(1,2,1)
cont = plt.contour(ww0,ww1,ff, 10, colors='b') #value  
cont.clabel(fmt='%2.0f', fontsize = 8)
plt.xticks(range (-w_range, w_range+1,1))
plt.yticks(range (-w_range, w_range+1,1))
plt.xlim(-w_range - 0.5,w_range + .5)
plt.ylim(-w_range - .5,w_range + .5)
plt.xlabel('$w_0$', fontsize = 14)
plt.ylabel('$w_1$', fontsize = 14)

plt.subplot(1,2,2)
plt.quiver(ww0, ww1,dff_dw0, dff_dw1) #f   
plt.xlabel('$w_0$', fontsize = 14)
plt.ylabel('$w_1$', fontsize = 14)
plt.xticks(range(-w_range, w_range+1,1))
plt.yticks(range(-w_range, w_range+1,1))
plt.xlim(-w_range - 0.5,w_range + .5)
plt.ylim(-w_range - .5,w_range + .5)
plt.show()
```

## non-linear activations

```python
import copy
def binary_step(_x):
    x = copy.deepcopy(_x)
    for i,v in enumerate(x): 
        if v>0 : x[i]=1.0
        if v<0 : x[i]=0.0
    return x
def sigmoid_function(x):
    z = (1/(1 + np.exp(-x)))
    return z    
def tanh_function(x):
    z = (2/(1 + np.exp(-2*x))) -1
    return z
def relu_function(_x):
    x = copy.deepcopy(_x)
    for i,v in enumerate(x): 
        if v>0 : x[i]=x[i]
        if v<0 : x[i]=0.0
    return x

def leaky_relu_function(_x,a=0.051):
    data = [max(a*value,value) for value in _x]
    return np.array(data, dtype=float)
def elu_function(_x, a):
    x = copy.deepcopy(_x)
    for i,v in enumerate(x): 
        if v>0 : x[i]=x[i]
        if v<0 : x[i]=a*(np.exp(x[i])-1)
    return x

def swish_function(_x):
    x = copy.deepcopy(_x)
    return x/(1-np.exp(-x))
def softmax_function(_x):
    x = copy.deepcopy(_x)
    z = np.exp(x)
    z_ = z/z.sum()
    return z_
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
plt.figure(figsize=(5,5))
plt.plot(x,binary_step(x),'gray', linestyle='--', linewidth=1, label='binary_step')
plt.plot(x,sigmoid_function(x),'black', linewidth=1, label='sigmoid')
plt.plot(x,tanh_function(x),'green', linewidth=1, label='tanh')
plt.plot(x,relu_function(x),'orange', linewidth=1, label='relu')
plt.plot(x,leaky_relu_function(x),'red', linewidth=1, label='leaky_relu')
plt.plot(x,elu_function(x,1),'blue', linestyle='-.', linewidth=1, label='elu')
plt.plot(x,swish_function(x),'cornflowerblue', linewidth=1, label='swish')
plt.plot(x,softmax_function(x),'violet', linewidth=1, label='softmax')

plt.xlim(-10,10)
plt.ylim(-1,5)
plt.grid(True)
plt.legend(loc='upper left')
```
