```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-4,4,100)
y1=2**x
y2=3**x
y3=0.5**x

print('x={}'.format(x))
print('y1={}'.format(y1))
print('y2={}'.format(y2))
print('y3={}'.format(y3))


plt.figure(figsize=(5,5))
plt.plot(x,y1, 'black', linewidth = 3, label = '$y=2^x$')
plt.plot(x,y2, 'cornflowerblue',linewidth=3, label = '$y=3^x$')
plt.plot(x,y3,'gray',linewidth =3, label = '$y=0.5^x$')
plt.ylim(-2,6)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='lower right')
plt.savefig('graph,png')
```

# 3차 함수 : f(x) = (x-2)x(x+2)

```python
import numpy as np
import matplotlib.pyplot as plt
#matplot inline

def f(x):
    y = (x-2)*x*(x+2)
    return y
print('x=1',f(1))
print('x=2',f(2))
print('x=3',f(3))

x = np.array([1,2,3])
x = np.arange (-3 , 3.5 , 0.5)
x = np.linspace (-3,3,100)
print('x[] = {} , type = {}'.format(x, type(x)))
print(f(x))


plt.figure(figsize=(5,5))
plt.plot(x,f(x),'cornflowerblue',linewidth=3, label = '$f(x):y = (x-2)*x*(x+2)$')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#matplot inline

def f(x):
    y = (x-2)*x*(x+2)
    return y

def f2(x,w):
    y = (x-w)*x*(x+2)
    return y

def f3(x,q):
    y = (x-q)*x*(x+2)
    return y

x = np.linspace (-3,3,50)
print('x[] = {} , type = {}'.format(x, type(x)))
print(f(x))


plt.figure(figsize=(5,5))
plt.plot(x,f(x),'cornflowerblue',linewidth=3, label = '$f(x):y = (x-2)*x*(x+2)$')
plt.plot(x,f2(x, w=4),'gray',linewidth=3, label = '$f(x):y = (x-4)*x*(x+2)$')
plt.plot(x,f3(x, q=2),'red',linewidth=3, label = '$f(x):y = (x-4)*x*(x+2)$')

plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.ylim(-30,30)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
plt.savefig('graph,png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
#matplot inline

def f3(x0,x1):
    r=2**x0**2 +x1**2
    ans = r*np.exp(-r)
    return ans

#x0, x1 에 따라서 각각 f3 함수를 계산
xn = 9
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((len(x0), len(x1)))#0으로 초기화
print('x0 = {}'.format(x0))
print('x1 = {}'.format(x1))
print('y = {}'.format(y))

for i0 in range(xn):
    for i1 in range(xn):
             y[i1,i0] = f3(x0[i0],x1[i1])
y = np.round(y,1)
print('y={}'.format(y))
plt.figure(figsize = (3.5,3))
#plt.gray()
#plt.jet()
plt.pink()
plt.pcolor(y)
plt.colorbar()
plt.savefig('gauss.png')
 
```

## 그래프 표현 - surface

```python
from mpl_toolkits.mplot3d import Axes3D

xn = 50
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
print(x0)
print(x1)

xx0, xx1 = np.meshgrid(x0,x1)
plt.figure(figsize = (5,3.5))
for i in range(10):
    ax = plt.subplot(4,4,i+1,projection='3d')
    ax.plot_surface(xx0,xx1,y,rstride = 1,cstride=1,alpha=0.5,
                    color = 'blue',edgecolor='black')

    ax.set_zticks((0,0.2))
    ax.view_init(40,-115)
```

## 그래프 표시 - contour

```python
from mpl_toolkits.mplot3d import Axes3D

xn = 50
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y= np.zeros((len(x0), len(x1)))

for row in range(xn):
    for col in range (xn):
        y[col,row] = f3 (x0[row],x1[col])
xx0, xx1 = np.meshgrid(x0,x1) # 4x4 행렬을 만들어주는 역할

plt.figure(figsize=(4,4))
cont = plt.contour(xx0,xx1,y,5,colors = 'black')
cont.clabel(fmt = '%3.2f',fontsize = 8)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)

```

```python
import numpy as np
import matplotlib.pyplot as plt
#matplot inline
def f(x,w):
    y = x*w
    return y

x=np.arange(10) 
x = np.linspace (0,10,50)
plt.figure(figsize = (5,5))
plt.plot(x,f(x,3),'black',linewidth=3, label = '$f(x):y = 3 * x$')
plt.plot(x,f(x,5),'red',linewidth=3, label = '$f(x):y = 5 * x$')
plt.plot(x,f(x,7),'green',linewidth=3, label = '$f(x):y = 7 * x$')
plt.plot(x,f(x,9),'blue',linewidth=3, label = '$f(x):y = 9 * x$')
plt.xlim(0,10)
plt.ylim(0,60)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
#matplot inline
def f(x,w):
    y = x*w
    return y

x=np.arange(10) 
plt.figure(figsize = (7,5))
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)

for i in range(4):
    plt.subplot(2,2,i+1) #plot 묘사의 위치 지정
    plt.title(i*2+3)
    plt.plot(x,f(x,i*2+1),'black')
    plt.xlim(0,10)
    plt.ylim(0,60)
    plt.grid(True)
plt.show()
```
